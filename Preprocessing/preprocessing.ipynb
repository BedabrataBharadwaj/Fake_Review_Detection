```python
import pandas as pd
import re
import string

df = pd.read_csv("dataset.csv")

df = df[['review_text', 'label', 'user_id', 'product_id', 'date']].dropna()

df = df.rename(columns={'product_id': 'review_id'})

behavior_df = df[['review_text', 'label', 'user_id', 'review_id', 'date']]

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+", "", text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

df['text'] = df['review_text'].apply(clean_text)

bert_df = df[['text', 'label']]

bert_df.to_csv("bert_input.csv", index=False)
behavior_df.to_csv("behavioral_input.csv", index=False)

print(bert_df.head())
```
